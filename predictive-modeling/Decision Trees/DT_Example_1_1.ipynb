### Decision Trees Example 1.1
We study the **Heart** data set. These data contain a binary outcome **AHD** for $303$ patients which experienced chest pain. The level **Yes** indicates the presence of a heart disease based on an angiographic test, while **No** means no heart disease. There are $13$ predictors including **Age**, **Sex**, **Chol** (cholesterol measurement), and other heart and lung function measurements. Some predictor values are shown below.

|| Sex | ChestPain | RestBP | Chol | Fbs | RestECG | MaxHR | ExAng|
|----|-----------|----------|---|------|---------|-------|-----|---|
|1 | 1 | typical | 145 | 233 | 1 | 2 | 150 | 0|
|2 | 1 | asymptomatic | 160 | 286 | 0 | 2 | 108 | 1|
|3 | 1 | asymptomatic | 120 | 229 | 0 | 2 | 129 | 1|
|4 | 1 | nonanginal | 130 | 250 | 0 | 0 | 187 | 0|
|5 | 0 | nontypical | 130 | 204 | 0 | 2 | 172 | 0|
|6 | 1 | nontypical | 120 | 236 | 0 | 0 | 178 | 0|

The challenge is to predict the presence of the heart disease by means of the predictor values only and without performing the angiographic test. The predictors **MaxHR** (maximal heart rate in bpm) and **Age** are depicted in a scatter plot where the colour corresponds to the value of the disease indicator **AHD**. 

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('./data/Heart.csv')

# Plot 
plt.figure(figsize=(7, 5))
sns.scatterplot(x=df['Age'], y=df['MaxHR'], 
                hue=df['AHD'], palette=['green', 'red'])
plt.show()
```

A separation of the two groups based on these two predictors becomes visible: Individuals that are younger and have a higher maximal heart rate seem to suffer less frequently from heart disease.
