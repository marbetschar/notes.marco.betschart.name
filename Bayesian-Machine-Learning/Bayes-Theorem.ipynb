# Bayes Theorem

:::{prf:definition} Bayesian Analysis, <doi:10.1201/9780429288340>
Every Bayesian analysis consists of four common steps.

1. Construct a **prior model** for your variable of interest, $\pi$. The prior model specifies two important pieces of information: the possible values of $\pi$ and the relative prior plausibility of each.

2. Summarize the dependence of data $Y$ on $\pi$ via a **conditional pmf** $f(y \mid \pi)$.

3. Upon *observing* data $Y = y$, define the **likelihood function** $L(\pi \mid y) = f(y \mid \pi)$ which encodes the relative likelihood of observing data $Y = y$ under different $\pi$ values.

4. Build the **posterior model** of $\pi$ via Bayesâ€™ Rule which balances the prior and likelihood:

$$
\text{posterior} = \dfrac{\text{prior} \cdot \text{likelihood}}{\text{normalizing constant}} \propto \text{prior} \cdot \text{likelihood}
$$

   More technically:

$$
f(\pi \mid y) = \dfrac{f(\pi)L(\pi \mid y)}{f(y)} \propto f(\pi)L(\pi \mid y)
$$
:::
